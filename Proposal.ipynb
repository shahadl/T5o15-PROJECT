# The Design

This project explores wine data to answer questions about what makes wine good quality based on physicochemical tests, and whether there are relationships between the variables or if there are differences in the effect of a variable according to the type of wine.

Finally, **is it possible to build a model that can predict with good accuracy the quality of wine?**

# The Dataset & it's Description

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine, and Some chemical properties ofvariables it.

**Input variables** (based on physicochemical tests):
1. fixed acidity
2. volatile acidity
3. citric acid
4. residual sugar
5. chlorides
6. free sulfur dioxide
7. total sulfur dioxide
8. density
9. pH
10. sulphates
11. alcohol

**Output variable** (based on sensory data):
12. quality (score between 0 and 10)

dataset url: `https://archive.ics.uci.edu/ml/datasets/wine+quality`
