# `MVP: Analysis of Census Dataset`

## This project to explore whether if demographic characteristics have an effect on an individual's annual income, and can we predict whether if the income is `<=50K` or `>50K`?

## I've got into all process of exploring the data and pre preparing it to fit a logistic regression model, then I calculated the accuracy and F-score of the prediction.

## I found the results promising to investigate further if I could improve this model or find another better machine learning method, perhaps Decision Trees or RandomForest.
